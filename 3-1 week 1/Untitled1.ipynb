```python
regions = [
    "과천시", "광명시", "광주시", "군포시", "김포시", "부천시", "성남시", "수원시", "시흥시", "안산시",
    "안성시", "안양시", "여주시", "오산시", "용인시", "의왕시", "이천시", "평택시", "하남시", "화성시", "양평군"
]

# '주소' 컬럼에서 경기 남부 지역이 포함된 데이터 필터링
pattern = "|".join([f"경기도 {region}" for region in regions])  # "경기도 과천시|경기도 광명시|..."
cctv_sn = cctv_sn[cctv_sn['소재지지번주소'].str.contains(pattern, na=False)]

cctv_sn.to_csv("경기남부_필터링.csv", index=False, encoding='utf-8-sig')
```
