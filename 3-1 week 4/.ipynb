```python
import json
with open('전국휴양림표준데이터.json','r') as files:
    temp = json.load(files)
temp
```

```python
data = []
for d in temp['records']:
    data.append(d)
data
```

```python
import pandas as pd
pdData = pd.DataFrame(data)
pdData.to_excel('휴양지.xlsx')
```

```python
df = pd.read_excel("휴양지.xlsx")
df.head()
```

```python
import folium
anMap = folium.Map(location=[37.4965,126.9573],
                   zoom_start=11,
                   zoom_control=False,
                   control_scale = True)
anMap
```

```python
for n, row in df.iterrows():
    print(n, row['휴양림명'], row['위도'], row['경도'])
    
```

```python
for n, row in df.iterrows():
    mk = folium.Marker(location=[row['위도'], row['경도']], popup = folium.Popup(row['휴양림명'], max_width=200))
    mk.add_to(anMap)
anMap
```
